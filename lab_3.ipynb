Посчитать теплозапас воды, т.е. определенныей интеграл от температуры по глубине (от поверхности до дна) по формулам:
- метод левых и правых прямоугольников
- метод трапеций
- метод Симпсона (метод парабол)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_file_path = "data/data.txt"
with open(data_file_path, "r") as f:
    df = pd.read_csv(f, sep=";")

df.head(n=10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

horizons = [5, 10, 20, 25, 30, 40]
temperatures_linear_interp = []
temperatures_cubic_interp = []

depth_column = "Depth,m"
temperature_column = "Temperature,C"

fig, ax = plt.subplots(1, 1)

ax.plot(df[temperature_column], df[depth_column], "r", label="Original Data")
ax.fill_betweenx(
    df[depth_column], df[temperature_column], color="b", alpha=0.3, label="Filled Area"
)

X = df[depth_column].tolist()
Y = df[temperature_column].tolist()

n = len(X) + 1
a, b = X[0], X[-1]

print("Water heat capacity by three methods:")

# Rectangle Method
h_rectangle = (b - a) / n
heat_capacity_rectangle = h_rectangle * sum(Y[:-1])
print(f"Rectangle Method: {heat_capacity_rectangle:.3f}")

# Trapezoidal Method
h_trapezoid = (b - a) / n
heat_capacity_trapezoid = h_trapezoid * ((Y[0] + Y[-1]) / 2 + sum(Y[1:-1]))
print(f"Trapezoidal Method: {heat_capacity_trapezoid:.3f}")

# Simpson's Rule
h_simpson = (b - a) / (3 * n)
heat_capacity_simpson = h_simpson * (Y[0] + 4 * sum(Y[1::2][:-1]) + 2 * sum(Y[2::2][:-1]) + Y[-1])
print(
    f"Simpson's Rule (Parabola): {heat_capacity_simpson:.3f}",
)

ax.set_title("Interpolation")
ax.invert_yaxis()
ax.legend()
plt.xlabel("Temperature, C")
plt.ylabel("Depth, m")
plt.show()
```
